```python

punctuation_chars = ["'", '"', ",", ".", "!", ":", ";", '#', '@']

############ Extracting Positive Words ##########################################

positive_words = []

with open("positive_words.txt") as pos_f:
    for lin in pos_f:
        if lin[0] != ';' and lin[0] != '\n':
            positive_words.append(lin.strip())

############# Extracting Negative Words to Use ##################################

negative_words = []
with open("negative_words.txt") as pos_f:
    for lin in pos_f:
        if lin[0] != ';' and lin[0] != '\n':
            negative_words.append(lin.strip())

################## Remove Garbage #################################################

def strip_punctuation(x):
    for ite in punctuation_chars:
        x=x.replace(ite,'')
        
        return x
    
##################### Get Negative Score #############################################
    
def get_neg(y):
    count=0
    y=y.split()
    
    for ite in y:
        ite=ite.lower()
        ite=strip_punctuation(ite)
        if ite in negative_words:
            count=count+1
            
    return count

######################## Get Positive Score ###########################################

def get_pos(y):
    count=0
    y=y.split()
    
    for ite in y:
        ite=ite.lower()
        ite=strip_punctuation(ite)
        if ite in positive_words:
            count=count+1
            
    return count

############################# Start of Main File ########################################

fileref = open("project_twitter_data.csv", "r")
fileref2= open("resulting_data.csv","w")

################################### Header Line #############################################

fileref2.write('Number of Retweets, Number of Replies, Positive Score, Negative Score, Net Score')
fileref2.write('\n')

################################### Main File #################################################

for aline in fileref.readlines()[1:]:
    aline=aline.strip()
    values = aline.split(',')
    
    row_string = '{},  {},  {}, {},  {}'.format(values[1], values[2],get_pos(aline),get_neg(aline),(get_pos(aline)-get_neg(aline)))
    fileref2.write(row_string)
    fileref2.write('\n')
    print(values)
    
 ################################# Close File after writing to them #################################

fileref.close()
fileref2.close()





```
